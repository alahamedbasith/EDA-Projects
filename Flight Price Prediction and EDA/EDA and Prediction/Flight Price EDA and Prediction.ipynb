#### Finally we are successfully completed the feature engineering

In most cases, you should not fill missing values in the dependent variable (also known as the target variable) when performing feature engineering in machine learning. The dependent variable is the variable you are trying to predict, and imputing or filling in missing values for it can lead to biased or incorrect model results.

Instead, you should focus on handling missing values in the independent variables (features) used for prediction.

## Model Building

- We are not fill the missing value in y because we have enough amount data.So just reshaping the feature X to Y.
- Suppose we fill the missing value the accuracy it may be small compare than reshaping.
- We have not so much data in y at that time,We need to fill the missing values.

We are does not do feature scaling because of the data is low magnitude.But feature scaling is Recommended for model building.

```python
from sklearn.model_selection import train_test_split

# Check for and fill missing values in y
missing_rows = df["Price"].isnull()

# Remove missing rows from the dataset with respect to price [Dependent variable]
df = df[~missing_rows]

# Separate X and y
X = df.drop("Price", axis=1).values
y = df["Price"].values

# Ensure that X has the same number of rows as y
X = X[:len(y), :]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

Here's a breakdown of the code X[:len(y), :]:

   - len(y) calculates the length (number of elements) in the array y. In the context of your code, it determines how many rows there should be in X to match the length of y.

   - : is a slicing operator in Python. When used with X[:len(y), :], it selects a range of rows and all columns in the array X. The colon : indicates that you want to include all columns.

So, X[:len(y), :] effectively means "Select all rows in X up to the length of y, and include all columns."

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor  # Example algorithm

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
